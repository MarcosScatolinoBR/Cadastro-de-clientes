Sistema de clientes.

Projeto de agenda digital com Python. A principio utilizarei csv ou excel para armazenar os dados de cadastro mas futuramente implementarei em sql.
Edit - o projeto cresceu de tal forma que ao inves de uma agenda mais simples acabei transformando em um sistema de clientes.

```python
'''
def cadastro ok
def consulta ok
def apagar ok
def editar
'''
import pandas as pd
import os.path
import datetime
import sys

#Verifica a hora do sistema apenas para fins de cortesia e saudação
hora = datetime.datetime.now()

def menu():
    while True:
        if 6 <= hora.hour < 12:
            print("Bom dia!")
        elif 12 <= hora.hour < 18:
            print("Boa tarde!")
        elif 18 <= hora.hour < 24:
            print("Boa noite!")
        else:
            print("Boa madrugada!")

        print('Escolha uma opção: ')
        print(' 1 - Cadastrar novo cliente')
        print(' 2 - Consultar cadastro')
        print(' 3 - sair')

        escolha = int(input())

        if escolha == 1:
            cadastro()
        elif escolha == 2:
            menu_consulta()
        elif escolha == 3:
            break

# Verificar se o arquivo clientes.csv já existe
if os.path.isfile('clientes.csv'):
    # Se o arquivo existir, carregar os dados existentes
    df_clientes = pd.read_csv('clientes.csv', encoding = 'utf-8', sep=',')
    print('Lista de clientes carregada com sucesso!')
else:
    # Se o arquivo não existir, carrega a lista de clientes
    df_clientes = pd.DataFrame(columns=['nome', 'cpf', 'telefone', 'email', 'endereco', 'cidade', 'estado'])
    df_clientes.to_csv('clientes.csv', index=False)
    print('Arquivo clientes.csv criado com sucesso!')

#Define a função de cadastro de usuários.
def cadastro():
    print('Cadastrar novo usuário')
    #verificação para saber se realmente deseja adiconar novo cadastro para evitar erros.
    print('Deseja realmente fazer um novo cadastro?')
    print('1 - Sim')
    print('2 - Não')
    escolha = int(input('Digite a opção desejada'))
    if escolha == 2:
        menu()
    elif escolha == 1:
        #formuário de cadastro.
        nome = input('Nome do cliente: ')
        cpf = input('CPF: ')
        telefone = input('Telefone cliente: ')
        email = input('E-mail cliente: ')
        endereco = input('Rua/Avenida: ')
        numero = input('Número: ')
        bairro = input('Bairro: ')
        cidade = input('Cidade: ')
        estado = input('Estado: ')
        novo_cliente = {
            'nome': nome,
            'cpf': cpf,
            'telefone': telefone,
            'email': email,
            'endereco': endereco,
            'número' : numero,
            'bairro': bairro,
            'cidade': cidade,
            'estado': estado,
        }
        #carrega o arquivo csv para a var df_clientes, decodifica para evitar erros e salva o novo cadastro no arquivo.
        df_clientes = pd.read_csv('clientes.csv', encoding='utf-8', sep=',')
        df_clientes = df_clientes.append(novo_cliente, ignore_index=True)
        df_clientes.to_csv('clientes.csv', index=False)
        print('Cliente cadastrado com sucesso!')

#menu da seção consulta
def menu_consulta():
    print('O que deseja fazer?')
    print('1 - Nova consulta')
    print('2 - Editar')
    print('3 - Apagar')
    print('4 - Voltar')
    print('5 - Sair')

    opcao = int(input('Digite a opção desejada: '))

    if opcao == 1:
       consulta()
    if opcao == 2:
       editar()
    if opcao == 3:
       apagar(df_clientes)
    if opcao == 4:
       menu()
    if opcao == 5:
       fim()

#seção de consulta de clientes. Verifica a existência de um dado X no arquivo csv.
def consulta():
    print('Escolha uma opção:')
    print('1 - Nome')
    print('2 - CPF')
    print('3 - Telefone')
    print('4 - Cidade')
    print('5 - Estado')
    print('6 - Menu')
    print('7 - Sair')
    busca = int(input('Digite a opção desejada: '))
    termo = str()
    if busca == 1:
        termo = 'nome'
    elif busca == 2:
        termo = 'cpf'
    elif busca == 3:
        termo = 'telefone'
    elif busca == 4:
        termo = 'cidade'
    elif busca == 5:
        termo = 'estado'
    elif busca == 6:
        menu()
    elif busca == 7:
        fim()
    else:
        print('Opção inválida.')
        return

    pesq = input(f'Entre com o {termo} para pesquisa: ').strip().lower()
    print("Pesquisa:", pesq)

    resultados = []

    df_clientes = pd.read_csv('clientes.csv', encoding='utf-8', sep=',')

    for index, row in df_clientes.iterrows():
        if pesq in str(row.loc[termo]).strip().lower():
            resultados.append(row)
            print("Termo de busca encontrado:", row[termo])

    if len(resultados) != 0:
        print('Resultados encontrados:')
        for resultado in resultados:
            print(resultado)
    else:
        print('Nenhum resultado encontrado.')

#Apagar registro de cliente.
def apagar(clientes):
    print('Você realmente deseja APAGAR TODOS OS DADOS de um cliente?')
    print('Após apagar não terá mais como recuperar esses dados!')
    print('1 - Sim')
    print('2 - Não')
    resposta = int(input('Digite 1 para sim e 2 para não'))
    if resposta == 1:
        print('IMPORTANTE! Você irá apagar dados de um cliente!')
        cpf = input("Entre com o CPF do cliente que deseja apagar: ")
        # Verifica se o CPF existe na lista de clientes
        print('Você realmente deseja apagar esses dados?')
        print('1 - Sim, apagar todos os dados!')
        print('2 - Não, não apagar dados!')
        verifica = int(input('Digite 1 para apagar os dados e 2 para não pagar e retornar ao menu: '))
        if verifica == 1:
            print('Apagando dados!')
            print('Eles não poderão mais ser recuperados ou acessados.')
            #analisa o dataframe df_clientes e busca por cpf para localizar o item desejado.
            for index, row in clientes.iterrows():
                # Convertendo o CPF para string antes de comparar
                if str(row['cpf']) == cpf:  
                    #Remove o cliente do df_clientes.
                    clientes = clientes.drop(index)
                      #Atualiza o arquivo CSV
                    clientes.to_csv('clientes.csv', index=False)
                    print("Cadastro apagado com sucesso!")
                    return
            #Caso não encontre o cpf, retorna a mensagem e retorna ao menu.
            print("CPF não encontrado. Nenhum cadastro foi apagado.")
            menu()
            
    elif resposta == 2:
        print('Nenhum dado foi apagado!')
        print('Retornando ao menu...')
        menu()
    else:
        print('Opção inválida, retornando ao menu...')
        menu()

def editar():
    print('Editar dados do usuário')
    print('Deseja realmente editar o cadastro?')
    print('1 - Sim')
    print('2 - Não')
    escolha = int(input('Digite a opção desejada: '))
    if escolha == 2:
        menu()
    elif escolha == 1:
        consulta_para_edicao()

def consulta_para_edicao():
    pesq = input('Entre com o CPF para edição: ').strip().lower()
    print("Pesquisa:", pesq)

    resultados = []

    df_clientes = pd.read_csv('clientes.csv', encoding='utf-8', sep=',')

    for index, row in df_clientes.iterrows():
        if pesq in str(row['cpf']).strip().lower():  # Modificado para pesquisar pelo CPF
            resultados.append(row)
            print("Termo de busca encontrado:", row['cpf'])  # Modificado para imprimir o CPF

    if len(resultados) != 0:
        print('Resultados encontrados:')
        for resultado in resultados:
            print(resultado)
    else:
        print('Nenhum resultado encontrado.')


def atualiza_csv():
    df_clientes = pd.DataFrame(clientes)
    df_clientes.to_csv('clientes.csv', index=False)

def fim():
    print('Encerrando programa...')
    return

menu()
```
